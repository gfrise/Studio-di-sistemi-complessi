```python
#Import third-party libraries
import numpy as np 
import matplotlib.pyplot as plt 

def logistic_map(r, y):
	'''a function to calculate the next step of the discrete map.  Inputs
	x and y are transformed to x_next, y_next respectively'''
	y_next = r * y * (1 - y)
	r_next = r + 0.0000001
	yield r_next, y_next
```

```python
steps = 30000000

Y = np.zeros(steps + 1)
R = np.zeros(steps + 1)

R[0], Y[0] = 1, 0.5

# map the equation to array step by step using the logistic_map function above
for i in range(steps):
	r_next, y_next = next(logistic_map(R[i], Y[i])) # calls the logistic_map function on R[i] as r and Y[i] as y
	R[i+1] = r_next
	Y[i+1] = y_next
```

```python
plt.figure(figsize=(10, 10))
plt.plot(R, Y, '^', color='red', alpha=0.4, markersize = 0.013)
plt.axis('on')
plt.show()
```
